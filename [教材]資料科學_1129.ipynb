# 遺失值(Missing Value)

## 遺失值的種類
1. 完全隨機缺失(missing completely at random, MCAR): 某一變數缺失值不依賴於其他任何原因的完全隨機缺失
2. 隨機缺失 (missing at random, MAR): 某一變數的缺失與其他變數相關但與該變數本身的數值不相關
3. 非隨機缺失(missing not at random, MNAR): 某一變數的缺失和該變數本身的數值相關， e.g., 儀器的最低檢測線：某被檢測物質的含量低於該檢測線則會產生非隨機缺失  
  
實務上MAR 較少見，故可將 NA 類型簡化為兩大類── MCAR 和 MNAR

## 數據缺失的處理方法
### 刪除記錄
* 優點：最簡單粗暴
* 缺點：犧牲了大量的數據，通過減少歷史數據換取完整的訊息，這樣可能丟失了很多隱藏的重要訊息；當缺失數據比例較大時，特別是缺失數據非隨機分佈時，直接刪除可能會導致數據發生偏離。

### 數據填補
* 用固定值填充
* 用均值填充
* 用中位數填充
* 用前一個數據進行填充
* 用後一個數據進行填充  
  
數據填補方式若不夠精準，可能會引入噪聲，或者會改變特徵原有的分佈

```python
iris = load_iris()
df_data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                       columns= ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'])
df_data
```

```python
# 檢查各欄位有無空值
df_data.isnull().sum() 
```

```python
# 無空值，故先隨機刪除部分的資料來模擬有缺失值的情況
sample = df_data.sample(50) # 抽樣50筆
random = np.random.randint(0,4,20) # 抽四個特徵中哪個要變nan

for i,j in enumerate(random): #列舉
    df_data.iloc[sample.index[i],j] = np.nan # 取代資料成nan
```

```python
# 計算各欄位空值
df_data.isnull().sum()
```

```python
# 描述統計
df_data.describe()
```

```python
# 用固定值填充
df_data['SepalLengthCm'] = df_data['SepalLengthCm'].fillna(value=0) 

# 用平均填充
df_data['SepalWidthCm'] = df_data['SepalWidthCm'].fillna(df_data['SepalWidthCm'].mean())

# 用中位數填充
df_data['PetalLengthCm'] = df_data['PetalLengthCm'].fillna(df_data['PetalLengthCm'].median())

# 用前一個數據進行填充
df_data['PetalWidthCm'] = df_data['PetalWidthCm'].fillna(method='pad')

# 用後一個數據進行填充
df_data['PetalWidthCm'] = df_data['PetalWidthCm'].fillna(method='bfill')

# 檢查有無空值
df_data.isnull().sum() 
```

# 離群值檢測(Outlier Detection)

指在數據中有一個或幾個數值與其他數值相比差異較大，以致影響到數值的分佈狀況

## 找出Outlier
* 查看每一個欄位的意義
* 檢查數值範圍(包含:平均數、標準差、中位數)
* 繪製散點圖(scatter)、分佈圖(histogram)、盒鬚圖(Box plot)等方式檢查

## 對 Outliers 的處理方法
* 刪除資料
* 保留資料：填補值(中位數、Min、Max、平均數)
* 另建特徵

```python
iris = load_iris()
df_data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                       columns= ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'])
```

```python
# 繪製盒鬚圖，檢測離群值
import matplotlib.pyplot as plt
for col in df_data.iloc[:,0:4]: #data.columns[0:4]
    plt.title(col)
    plt.boxplot(x=df_data[col], sym ="o", whis = 1.5)
    plt.show()
```

```python
print('第25百分位數：',df_data['SepalWidthCm'].describe()['25%']) # 查看第25個百分位數
print('第50百分位數：',df_data['SepalWidthCm'].describe()['50%']) # 查看第50個百分位數
print('第75百分位數：',df_data['SepalWidthCm'].describe()['75%']) # 查看第75個百分位數
```

## 處理SepalWidthCm的離群值
- 離群值定義：(Q3＋1.5\*IQR, Q1－1.5\*IQR)

```python
# 計算IQR
IQR = df_data['SepalWidthCm'].describe()['75%'] - df_data['SepalWidthCm'].describe()['25%']
IQR
```

```python
upper_bound = df_data['SepalWidthCm'].describe()['75%'] + (1.5*IQR)
lower_bound = df_data['SepalWidthCm'].describe()['25%'] - (1.5*IQR)
print('upper_bound: ',upper_bound)
print('lower_bound: ',lower_bound)
```

```python
df_data['SepalWidthCm'][df_data['SepalWidthCm'] > upper_bound] = df_data['SepalWidthCm'].describe()['75%']
```

```python
df_data['SepalWidthCm'][df_data['SepalWidthCm'] < lower_bound] = df_data['SepalWidthCm'].describe()['25%']
```

```python
# 繪製盒鬚圖，檢測離群值
import matplotlib.pyplot as plt
for col in df_data.iloc[:,0:4]: #data.columns[0:4]
    plt.title(col)
    plt.boxplot(x=df_data[col], sym ="o", whis = 1.5)
    plt.show()
```

# 正規化(Normalization) / 標準化(Standardization)

## 正規化(又稱歸一化)
* x' = (x - x_min) / (x_max - x_min)
* 將數據縮放到(0,1)之間
* 將有量綱化為無量綱表達式，成為純量，有便於不同單位或量級的指標能夠進行比較

## 標準化
* x' = (x - μ)／σ
* 在機器學習中，我們可能要處理不同種類的資料，例如，音訊和圖片上像素值等高維度資料，資料標準化後會使每個特徵中的數值平均變為0(將每個特徵的值都減掉原始資料中該特徵的平均)、標準差變為1，
* 這個方法被廣泛的使用在許多機器學習算法中(例如：支持向量機、邏輯回歸和類神經網絡)

## 功能
* 特徵間的單位（尺度)不同，比如身高和體重、攝氏度和華氏度、房屋面積和房間數，需消除特徵間單位和尺度差異的影響，以對每維特徵同等看待，需要對特徵進行正規化
* 正規化/標準化是一種線性變換，線性變換不會改變原始數據的數值排序，反而能提高數據表現
* 在使用梯度下降的方法求解最優化問題時， 歸一化/標準化後可以加快梯度下降的求解速度，即提升模型的收斂速度

## 正規化 / 標準化使用時機
* 如果對輸出結果範圍有要求，用正規化。
* 如果數據較為穩定，不存在極端的最大最小值，用正規化。
* 如果數據存在異常值和較多噪音，使用標準化。

```python
from sklearn import preprocessing

def normalize(x_train, method = 'MinMaxScaler'):
    '''
    method = 'StandardScaler', 'MinMaxScaler'
    '''    
    if method == 'StandardScaler':
        normalize_function = preprocessing.StandardScaler()
    if method == 'MinMaxScaler':
        normalize_function = preprocessing.MinMaxScaler()
    
    features = x_train.columns
    for feature in features:
        reshape = np.array(x_train[feature]).reshape(-1, 1)
        x_train[feature] = normalize_function.fit_transform(reshape)

    return x_train
```

```python
def plot_CM(true_label, pred_label, title = ' '):
    
    # Creates a confusion matrix
    cm = confusion_matrix(true_label, pred_label) 

    # Transform to df for easier plotting
    cm_df = pd.DataFrame(cm,
                         index = ['0', '1', '2'], 
                         columns = ['0', '1', '2'])

    plt.figure(figsize=(5.5,4))
    sns.heatmap(cm_df, annot=True, fmt="d")
    acc = accuracy_score(true_label, pred_label)

    recall = recall_score(true_label, pred_label, average='micro')
    precision = precision_score(true_label, pred_label, average='micro')

    plt.title(title + '\nAccuracy:{0:.3f}\nRecall:{1:.3f}\nPrecision:{2:.3f}'.format(acc, recall, precision))
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
```

### Macro average & Micro average
- 多元分類中，要綜合考量評價指標時候就會用到Macro average和Micro average的概念

```python
# Macro average recall 
Macro_average_recall = recall_score(predict, y_test, average = 'macro')
print('Macro average recall: ', Macro_average_recall)

# Micro average recall 
Micro_average_recall = recall_score(predict, y_test, average = 'micro')
print('Micro average recall : ', Micro_average_recall)

# Macro average precision
Macro_average_precision = precision_score(predict, y_test, average = 'macro')
print('Macro average precision: ', Macro_average_precision)

# Micro average precision
Micro_average_precision = precision_score(predict, y_test, average = 'micro')
print('Micro average precision : ', Micro_average_precision)
```

# 交叉驗證(Cross Validation)

一般來說我們在做機器學習 ( Machine Learning ) 的時候，我們會習慣將資料集 ( Dataset ) 切割 ( Splits ) 成訓練集 ( Training Set ) 跟驗證集 ( Validation Set )。顧名思義，Training Set是用來訓練 Machine Learning 的 Model；而 Validation Set 則用來驗證這個 Model 訓練的好不好。

K-fold Cross-Validation 將資料集拆分成 K 份，分別做training和validating，其中 K-1 份的資料當作訓練集，剩下來的那份做為驗證集，算出一個評價指標，接著再從沒當過驗證集的資料挑一份出來當驗證集，剛剛做過驗證集的資料則加回訓練集，維持 K-1 份做訓練、1 份做驗證，如此反覆直到每一份資料都當過驗證集，反覆執行 K 次，算出 K 個評價指標，最後我們再將這 K 個評價指標做平均，用他們的平均分數來做為我們評斷模型好壞的指標。

下載 (6).png

```python
iris = load_iris()
df_data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                       columns= ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'])
# 自變數、應變數
X = df_data.drop('Species', axis = 1)
y = df_data['Species']

# 切分訓練集、測試集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

X_train_norm = normalize(X_train)
X_test_norm = normalize(X_test)

#使用上面整理好切割好的資料做訓練
lr = LogisticRegression()
model = lr.fit(X_train_norm,y_train)

#預測分類結果
predict = model.predict(X_test_norm)
```

```python
# 進行交叉驗證
scores = cross_val_score(lr, X_train, y_train, cv=5, scoring = 'accuracy')
scores
```

## 特徵選擇(Feature Selection)
- 特徵選擇是指從眾多可用的特徵中選擇一個子集的過程，其目的和預期效果包含縮短訓練時間、增加模型效能、改善通用性與降低過擬合等

參考資料：https://scikit-learn.org/stable/modules/feature_selection.html

```python
iris = load_iris()
df_data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                       columns= ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'])

# 自變數、應變數
X = df_data.drop('Species', axis = 1)
y = df_data['Species']
```

### 變異數分析

```python
from sklearn.feature_selection import VarianceThreshold

#方差選擇法，返回值為特徵選擇後的數據，threshold為方差的閾值
result = VarianceThreshold(threshold=0.5).fit_transform(X)
result
```

```python
print('SepalLengthCm變異', np.var(X['SepalLengthCm']))
print('SepalWidthCm變異', np.var(X['SepalWidthCm']))
print('PetalLengthCm變異', np.var(X['PetalLengthCm']))
print('PetalWidthCm變異', np.var(X['PetalWidthCm']))
```

### 卡方分析
- 檢驗定性自變量對定性因變量的相關性

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

result = SelectKBest(chi2, k=3).fit_transform(X, y)
result
```

### LASSO

- 一個可同時進行特徵選擇和正則化的迴歸分析方法，旨在增強統計模型的預測準確性和可解釋性，最初用於計算最小平方法模型
- 線性模型中，可使用Lasso(Least Absolute Shrinkage and Selection Operator) 搭配 L1 regularization 讓不重要的係數值變成0，留下重要的特徵

 - L1 regularization
 - 模型中的變數需能使誤差平方和(SSE)降低對應水準，該特徵係數才會上升
 - 透過λ控制懲罰參數的強度
 - Loss: 
下載 (7).png

```python
from sklearn.linear_model import Lasso

reg = Lasso(alpha=0.1) # 調整 alpha 值決定正則化的強度
reg.fit(X, y)

coef=pd.Series(reg.coef_,index=X.columns)
coef
```

### Ridge

- 線性模型中，可使用Ridge Regression 搭配 L2 regularization 讓不重要的係數值變成0，留下重要的特徵

 - L2 regularization
 - 對係數做出二階懲罰
 - 透過λ控制懲罰參數的強度
 - Loss: 
下載 (8).png

```python
from sklearn.linear_model import Ridge 

reg = Ridge(alpha=0.1) #調整alpha值決定正則化的強度
reg.fit(X, y) 

coef=pd.Series(reg.coef_,index=X.columns)
coef
```

## 降維

### PCA
- 主成分分析(Principal components analysis)
- 線性降維

```python
iris = load_iris()
df_data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                       columns= ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'])

# 自變數、應變數
X = df_data.drop('Species', axis = 1)
y = df_data['Species']
```

```python
from sklearn.decomposition import PCA

X = normalize(X)

#主成分分析法，返回降維後的數據，參數n_components為主成分數目
result = PCA(n_components=2).fit_transform(X)
result
```

```python
# PCA結果視覺化
result_df = pd.DataFrame(result, columns = ['PCA1', 'PCA2'])
final_df = pd.concat([result_df,y], axis = 1)
final_df
```

```python
sns.lmplot("PCA1", "PCA2", hue='Species', data=final_df, fit_reg=False, legend=False)
plt.legend(title='Species', loc='upper right', labels=['Iris-Setosa', 'Iris-Versicolour', 'Iris-Virginica'])
```

### t-SNE
- t-Distributed Stochastic Neighbor Embedding (t-SNE)
- 非線性的機器學習降維方法

```python
from sklearn import manifold, datasets

#t-SNE
X_tsne = manifold.TSNE(n_components=2, init='random', random_state=5, verbose=1).fit_transform(X)
```

```python
# PCA結果視覺化
result_df = pd.DataFrame(X_tsne, columns = ['tSNE1', 'tSNE2'])
final_df = pd.concat([result_df,y], axis = 1)
final_df
```

```python
sns.lmplot("tSNE1", "tSNE2", hue='Species', data=final_df, fit_reg=False, legend=False)
plt.legend(title='Species', loc='upper right', labels=['Iris-Setosa', 'Iris-Versicolour', 'Iris-Virginica'])
```

# 超參數調整

- 以SVM分類模型為例

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, accuracy_score

iris = load_iris()
df_data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                       columns= ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'])

# 自變數、應變數
X = df_data.drop('Species', axis = 1)
y = df_data['Species']
```

```python
SVM = SVC()
SVM.__dict__
```

```python
# GridSearchCV是建立一個dictionary來組合要測試的參數
# C: Regularization parameter
# gamma: Kernel coefficient
# Kernel type to be used in the algorithm
param_grid = {'C':[1, 2, 5, 10, 50],
              'gamma':[1, 0.1, 0.01 ,0.001 ,0.0001]}
```

```python
grid = GridSearchCV(SVM, param_grid)
```

```python
#利用剛剛設定的參數來找到最適合的模型
final_model = grid.fit(X, y)
```

```python
#顯示最佳estimator參數
grid.best_estimator_
```
