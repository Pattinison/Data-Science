# 延伸練習6

```python
# 1. 將SepalWidthCm中的離群值資料丟棄
iris = load_iris()
df_data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                       columns= ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'])

# 計算IQR
IQR = df_data['SepalWidthCm'].describe()['75%'] - df_data['SepalWidthCm'].describe()['25%']
IQR

upper_bound = df_data['SepalWidthCm'].describe()['75%'] + (1.5*IQR)
lower_bound = df_data['SepalWidthCm'].describe()['25%'] - (1.5*IQR)
print('upper_bound: ',upper_bound)
print('lower_bound: ',lower_bound)

### start your code ###
# 1. 將SepalWidthCm中的離群值資料丟棄
drop_index = df_data['SepalWidthCm'][(df_data['SepalWidthCm'] > upper_bound) | (df_data['SepalWidthCm'] < lower_bound)].index
df_data['SepalWidthCm'].drop(index = drop_index)

# 2. 將SepalWidthCm中的離群值以平均數取代
iris = load_iris()
df_data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                       columns= ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'])

### start your code ###
df_data['SepalWidthCm'][df_data['SepalWidthCm'] > upper_bound] = df_data['SepalWidthCm'].mean()
df_data['SepalWidthCm'][df_data['SepalWidthCm'] < lower_bound] = df_data['SepalWidthCm'].mean()

# 3. 放寬/加嚴離群值判定邊界
### start your code ###
upper_bound = df_data['SepalWidthCm'].describe()['75%'] + (2*IQR)
lower_bound = df_data['SepalWidthCm'].describe()['25%'] - (2*IQR)
print('upper_bound: ',upper_bound)
print('lower_bound: ',lower_bound)
```

# 延伸練習7

```python
from sklearn import preprocessing

def normalize(x_train, method = 'MinMaxScaler'):
    '''
    method = 'StandardScaler', 'MinMaxScaler'
    '''    
    if method == 'StandardScaler':
        normalize_function = preprocessing.StandardScaler()
    if method == 'MinMaxScaler':
        normalize_function = preprocessing.MinMaxScaler()
    
    features = x_train.columns
    for feature in features:
        reshape = np.array(x_train[feature]).reshape(-1, 1)
        x_train[feature] = normalize_function.fit_transform(reshape)

    return x_train
```

```python
iris = load_iris()
df_data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                       columns= ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'])
df_data

### start your code ###
df_data = normalize(df_data)
df_data
```

# 延伸練習 8

```python
from imblearn.over_sampling import SMOTE 

iris = load_iris()
df_data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                       columns= ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'])
df_data

# 取前110筆，造成資料不平衡
df_data = df_data.head(110)
df_data

X = df_data.drop('Species', axis = 1)
y = df_data['Species']

print('Class 0 num: ', sum(y == 0.0))
print('Class 1 num: ', sum(y == 1.0))
print('Class 2 num: ', sum(y == 2.0))

### start your code ###
sm = SMOTE(random_state=1)
X_sm, y_sm = sm.fit_resample(X, y)

print('Class 0 num: ', sum(y_sm == 0.0))
print('Class 1 num: ', sum(y_sm == 1.0))
print('Class 2 num: ', sum(y_sm == 2.0))
```

# 延伸練習9

```python
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from xgboost.sklearn import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

iris = load_iris()
df_data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                       columns= ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'])

# 自變數、應變數
X = df_data.drop('Species', axis = 1)
y = df_data['Species']

# 切分訓練集、測試集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print('X_train shape:',X_train.shape)
print('X_test shape:',X_test.shape)
print('y_train shape:',y_train.shape)
print('y_test shape:',y_test.shape)

### start your code ###
# 1.DecisionTreeClassifier
DT = tree.DecisionTreeClassifier()
DT.fit(X_train, y_train)
DT_pred = DT.predict(X_test)
DT_acc = accuracy_score(DT_pred, y_test)
print('DT acc:', DT_acc)

# 2.RandomForest
RF = RandomForestClassifier(max_depth=2, random_state=0)
RF.fit(X_train, y_train)
RF_pred = RF.predict(X_test)
RF_acc = accuracy_score(RF_pred, y_test)
print('RF acc:', RF_acc)

# 3.XGBClassifier 
XGB = XGBClassifier()
XGB.fit(X_train, y_train)
XGB_pred = XGB.predict(X_test)
XGB_acc = accuracy_score(XGB_pred, y_test)
print('XGB acc:', XGB_acc)
```
